```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# Import the sqlalchemy extract function.
from sqlalchemy import extract

# 1. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temps = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 6)
```

```python
# 2. Convert the June temperatures to a list.
june_temps = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()

# Confirming that june_temps is a list
print(type(june_temps))
print(june_temps)
```

```python
# 3. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_temps, columns = ["June Temps"])
display(june_df.head(10))
```

```python
# 4. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

## Deliverable 3 Additional Queries

### Query 1: Determining if June or December has more days with precipitation

```python
# Finding the precipitation measurements for each day in June across all stations
june_prcp = session.query(Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()

# Finding the precipitation measurements for each day in December across all stations
december_prcp = session.query(Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()
```

```python
# Converting each of the above lists into DataFrames
june_prcp_df = pd.DataFrame(june_prcp, columns = ["June Precipitation"])
december_prcp_df = pd.DataFrame(december_prcp, columns = ["December Precipitation"])
```

```python
# Presenting the summary statistics for the June precipitation dataframe
display(june_prcp_df.describe())
```

```python
# Presenting the summary statistics for the December precipitation dataframe
display(december_prcp_df.describe())
```

### Query 2: Determining the summary statistics for the precipitation data gathered by each station in the months of June and December

```python
# Finding the summary statistics of the precipitation observations per station for the month of June

june_prcp_per_station = session.query(Measurement.station, func.min(Measurement.prcp), func.max(Measurement.prcp), 
                                     func.avg(Measurement.prcp), func.count(Measurement.station)).\
                                     group_by(Measurement.station).filter(extract('month', Measurement.date) == 6).\
                                     order_by(func.count(Measurement.station).desc()).all()

june_prcp_per_station
```

```python
# Converting the above data for June precipitation per station into a DataFrame
june_prcp_station_df = pd.DataFrame(june_prcp_per_station, 
                                   columns = ["Station", "Min. Precipitation", "Max. Precipitation", 
                                              "Avg. Precipitation", "Recorded Days with Precipitation"])

june_prcp_station_df
```

```python
# Finding the summary statistics of the precipitation observations per station for the month of December

dec_prcp_per_station = session.query(Measurement.station, func.min(Measurement.prcp), func.max(Measurement.prcp), 
                                     func.avg(Measurement.prcp), func.count(Measurement.station)).\
                                     group_by(Measurement.station).filter(extract('month', Measurement.date) == 12).\
                                     order_by(func.count(Measurement.station).desc()).all()

dec_prcp_per_station
```

```python
# Converting the above precipitation data per station for December into a DataFrame

dec_prcp_station_df = pd.DataFrame(dec_prcp_per_station, 
                                   columns = ["Station", "Min. Precipitation", "Max. Precipitation", 
                                              "Avg. Precipitation", "Recorded Days with Precipitation"])

dec_prcp_station_df
```
